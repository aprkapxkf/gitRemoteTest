```python
# 1번 문제
# 어떤 설문조사 시스템에서 나이를 입력받았습니다. 입력받은 age_input = "25"를 이용해 10년 뒤의 나이를 계산하여 "10년 뒤 당신은 35세가 됩니다."라고 출력하는 코드를 작성하세요.
def age() : 
    age_input = int(input("나이를 입력해주세요 : "))
    age_input = age_input+10
    print(f"당신의 10년 후 나이는 {age_input} 입니다.")
age()
```

```python
# 2번 문제
# 다음 리스트에서 중복된 과일 이름을 제거하고, 가나다 순(오름차순)으로 정렬된 리스트를 출력하세요.
def sort1(list):
    a = set(list)
    b = sorted(a)
    print(b)
fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]
sort1(fruits)
```

```python
# 3번 문제
# 아래와 같은 성적표 딕셔너리가 있습니다. 사용자가 "과학" 과목의 점수를 조회하려고 할 때, 에러가 발생하지 않도록 하고 데이터가 없을 경우 "해당 과목의 점수가 없습니다."라고 출력하는 코드를 작성하세요.
scores = {"국어": 90, "수학": 85, "영어": 95}
k = input("점수를 확인할 과목을 입력해주세요 : ") 
result = scores.get(k, "해당 과목의 점수가 없습니다.")
print(result)
```

```python
# 5번 문제
# 사용자로부터 입력받은 검색어 `search_keyword`가 비어있는지 확인하는 코드를 작성하세요.
#- 검색어가 있으면: **" '검색어'를 검색합니다."** 출력
#- 검색어가 비어있거나 공백만 있다면(strip 활용): **"검색어를 입력해주세요."** 출력
def search():
    search_keyword = input("키워드를 입력해주세요")
    search_keyword = search_keyword.strip()
    if len(search_keyword) == 0:
        print("검색어를 입력해주세요.")
    else:
        print(f"{search_keyword}를 검색합니다.")
search()
        
```

```python
# 10번 문제
# 웹사이트 방문자 데이터가 [(날짜, 방문자수), ...] 형태의 튜플 리스트로 저장되어 있습니다. 이 데이터를 시각화하기 위해 날짜 리스트(dates)와 방문자수 리스트(counts)로 각각 분리(Unzipping)한 뒤, 방문자 수의 평균을 계산하세요.

visitor_data = [("2026-03-01", 120), ("2026-03-02", 150), ("2026-03-03", 90), ("2026-03-04", 200)]

dates, count = zip(*visitor_data)
dates = list(dates)
count = list(count)

average = 0
for i in count:
    average += i
average = average/len(count)

print(f"날짜 리스트 : {dates}")
print(f"방문자수 리스트 : {count}")
print(f"방문자수 평균 : {average}")
```
